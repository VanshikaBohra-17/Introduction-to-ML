Making a column as index

```python
pd.read_csv('/content/aug_train.csv',index_col='enrollee_id')
```

When you want limited cols only - usecols

```python
pd.read_csv('/content/aug_train.csv',usecols=['city','gender','major_discipline'])
```

When you're working with only one column and want it as pandas series instead of pandas dataframe - squeeze parameter

```python
from numpy import squeeze
pd.read_csv('/content/aug_train.csv', usecols=['gender'])['gender']
#making Nan as 'notspecified'
```

skipping rows intentionally based on their index/sno. - 0 based - skiprows parameter

```python
pd.read_csv('/content/aug_train.csv',skiprows=[0,10])
```

getting only some rows or a batch of rows starting from 0 - nrows

```python
pd.read_csv('/content/aug_train.csv', nrows=10)
```
